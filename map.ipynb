```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

import folium
import contextily as ctx
```

```python
limits = [
    (54, 12),
    (54, 17),
    (56, 17),
    (56, 12),
    (54, 12),
]

m2 = folium.Map(
    location=(55, 12), 
    zoom_start=7,
    tiles=ctx.providers.OpenStreetMap.Mapnik,
    zoom_control=False
)

folium.Polygon(
    limits, 
    color = 'blue',
    fill_color="blue", 
    opacity=0.05,
    fill_opacity=0.3,
).add_to(m2)

folium.Polygon(
    limits,
    color = 'blue',
    fill=None
).add_to(m2)

m2
```
